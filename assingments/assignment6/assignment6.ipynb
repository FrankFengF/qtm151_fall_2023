(a) Compute function for each individual (map)

<font size = "3">

- Import the datset "data/bmi.csv"
    - **Note that the column `Height` is in $cm$, and `Weight` is in $kg$**
- Write a lambda function for $ BMI = \frac{weight (kg)}{height (m)^2} $
- Use ```list(map(...))``` to calculate $BMI$ for each row of the dataset
- Round the outcome into 2 decimal points
- Store the calculation in a new column in the dataframe called `BMI`
- Display the dataframe


````python
# Write your own code here
bmi=pd.read_csv("data/bmi.csv")
BMI=lambda w,h: round(w/((h/100)**2),2)
bmi['BMI']=list(map(BMI,bmi['Weight'],bmi['Height']))
display(bmi)
````

- Print the original `countryname` and the newly created column, like: <br>
"['Afghanistan', 'Africa Eastern and Southern', 'Africa Western and Central', 'Albania', 'Algeria', ......] "

````python
# Write your own code here
country=pd.read_csv("data/wdi_2020.csv")
fn_upper=lambda x: x.upper()
country['countryname_upper']=list(map(fn_upper,country['countryname']))
name_old=country['countryname'].tolist()
name_new=country['countryname_upper'].tolist()
print(name_old)
print(name_new)
````

(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code here
urban=pd.read_csv("data/wdi_urban.csv")
display(urban)
list_unique_year = pd.unique(urban["year"])
for i in list_unique_year:
    df   = urban.query("year == @i")
    plt.hist(x = df["prop_urbanpopulation"],alpha=0.5)
plt.xlabel("Prop_urbanpopulation")
plt.ylabel("Frequency")
plt.legend(labels = list_unique_year,
           title  = "Year")
plt.title('Histogram of proportion of urbanpopulation')
plt.show()
#We know that the proportion of urban population is higher in 2020 than 1980
````
